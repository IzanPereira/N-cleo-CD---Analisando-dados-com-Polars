```python
# Importando a biblioteca polars
import polars as pl
```

```python
#Lendo a o dataset winemag-data_first150k
wine = pl.read_csv('winemag-data_first150k.csv')
```

```python
nova_wine = wine.drop_nulls() # aqui estou instanciando um novo dataframe chamado nova_wine
                              # ja com as linha que contém dados nulos excluidos.
```

```python
nova_wine.describe() # verificando novamente as métricas principais como describe()
                      # verificando se esta tudo certo, agora não temos mais dados nulos
                      #como pode ver em count todos os valores estão iguais
                      # com a limpeza restaram 39241 linhas
```

```python
wine['country'].unique() # aqui eu só estou verificando os valores unico
```

```python
wine.filter(pl.col("country") == "Brazil").select(pl.col("variety").unique()) #Aqui estou só filtrando as variedades
                                                                              # que estão no Brasil um total de 8
```

```python
wine.filter((pl.col("country") == "Brazil") # aqui eu filtro por país(country)
                  & (pl.col("points") >= 90) # aqui eu filtro por nota(>=90)
                          & (pl.col("price") < 20)).select(pl.col("variety")) #aqui eu filtro por preço menor 20
                                                    #tudo isso selecionado a variedade
                                                    #porém como não obteve resultado devido a condição não ser satisfeita
```

```python
wine.filter((pl.col("country") == "Brazil")
                  & (pl.col("points") >= 85)
                          & (pl.col("price") < 20)).select(pl.col("variety"))
#porém quando mudo o condição para 85 pontos, vemos que o Merlot, podemos manipular os valores para
#encontrar uma lista maior, que atenda ao requesitos
```

```python
Agrupamento = wine.group_by("country", maintain_order=True).all() #aqui só estou agrupando
                                                    # agora temos 49 paíse no total
```

```python
print(Agrupamento)
```

```python
preçomedio = wine.group_by("country").agg(pl.col("price").mean().alias("price-medio")).sort("price-medio", descending=True).drop_nulls("price-medio")
#estou agrupando
#calculando a média
#colocando dando um nome para a coluna da média
#colocando na ordem do maior para o menor
```

```python
print(preçomedio) #US-France possui o maior preço na media
```

```python
nova_wine.group_by("country").agg(pl.col("price").mean().alias("price-medio")).sort("price-medio", descending=True)
#só estou verificando no dataframe limpo, não a outros dados a não ser do US
#dutante a limpeza dos valores em branco, todos os outros paises foram esxluidos
```

```python
notamax = wine.group_by("variety").agg(pl.col("points").max().alias("points-max")).sort("points-max", descending=True)
#estou agrupando
#calculando a maxima
#colocando dando um nome para a coluna da maxima
#colocando na ordem do maior para o menor
```

```python
print(notamax)
```

```python
nota_chardonnay = (wine.filter(pl.col("variety") == "Chardonnay").group_by("variety").agg(pl.col("points").max().alias("points-max"))
)
#estou filtrando por variedade por grupo e nota maxima
```

```python
print(nota_chardonnay)
```

```python
pontuações_chardonnays = (wine.filter(pl.col("variety") == "Chardonnay").filter(pl.col("points") == pl.col("points").max()))
#aqui estou filtrando e calculando os valores de todo o dataset, trazendo o país, a notra e o valor
```

```python
print(pontuações_chardonnays['country','points', 'price'])
```

```python
precos_medios = (wine.group_by("variety").agg(pl.col("price").mean().alias("preco_medio")).sort("preco_medio", descending=True)).drop_nulls("preco_medio")
#aqui estou filtrando e calculando os valores de todo o dataset e tirando os valores vazios
```

```python
print(precos_medios)
```

```python
wine = wine.with_columns((pl.col("points") / pl.col("price")).alias("custo-beneficio"))
# aqui estou criando a coluna com with_columns()
# já passando o argumento que sera points/price
```

```python
print(wine['points', 'price', 'custo-beneficio'])
```
